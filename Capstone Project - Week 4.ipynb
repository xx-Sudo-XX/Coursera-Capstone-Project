In groups we again have several different items, where type and name are irrelevant for given task. The full path to items is data.json()["response"]["groups"][0]["items"].

```python
items = d["groups"][0]["items"]
print("number of items: %i" % len(items))
items[0]
```

```python
items[1]
```

```python
df_raw = []
for item in items:
    venue = item["venue"]
    categories, uid, name, location = venue["categories"], venue["id"], venue["name"], venue["location"]
    assert len(categories) == 1
    shortname = categories[0]["shortName"]
    if not "postalCode" in location:
        continue
    postalcode = location["postalCode"]
    lat = location["lat"]
    lng = location["lng"]
    datarow = (uid, name, shortname, postalcode, lat, lng)
    df_raw.append(datarow)
df = pd.DataFrame(df_raw, columns=["uid", "name", "shortname", "postalcode", "lat", "lng"])
print("found %i cafes" % len(df))
df.head()
```

```python
ny_center = d["geocode"]["center"]
ny_center
```

Using a density based estimator will give us insight into where to start a new coffee shop. There's a HeatMap plugin ready in Folium, let's use that, and visualize all the existing Cafes to same map:

```python
from folium import plugins

# create map of Helsinki using latitude and longitude values
map_ny = folium.Map(location=[ny_center["lat"], ny_center["lng"]], zoom_start=14)

def add_markers(df):
    for (j, row) in df.iterrows():
        label = folium.Popup(row["name"], parse_html=True)
        folium.CircleMarker(
            [row["lat"], row["lng"]],
            radius=5,
            popup=label,
            color='white',
            fill=True,
            fill_color='#500cc',
            fill_opacity=0.7,
            parse_html=False).add_to(map_ny)

add_markers(df)
hm_data = df[["lat", "lng"]].as_matrix().tolist()
map_ny.add_child(plugins.HeatMap(hm_data))

map_ny
```

```python
lat = 40.742185
lng = -73.992602
map_ny = folium.Map(location=[lat, lng], zoom_start=17)
add_markers(df)
folium.CircleMarker(
    [lat, lng],
    radius=15,
    popup="Our Cafe!",
    color='red',
    fill=True,
    fill_color='#3186cc',
    fill_opacity=0.7,
    parse_html=False).add_to(map_ny)
map_ny
```
